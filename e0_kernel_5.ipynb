```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib.pyplot import figure
import scipy.stats as stat
```

```python
# Import and clean data
df = pd.read_csv('df_kernel_5.csv')
df.drop(columns = 'Unnamed: 0', inplace = True)
df[::50]
```

```python
# Plot
figure(figsize = (35, 15))
sns.set_style("darkgrid")

plt.set_cmap("gray")
# plt.subplot(121)

plt.subplot(121)
sns.scatterplot(x = 'Activity', y = 'Correlation', data = df)
plt.xscale('log')

plt.subplot(122)
sns.scatterplot(x = 'Activity', y = 'Detection', data = df)
plt.xscale('log')
```

```python
# Calculate correlation and significance level
correlation, p_value = stat.pearsonr(df['Detection'], df['Activity'])
print(f"Correlation: {round(correlation, 3)}")
print(f"P Value: {round(p_value, 6)}")
```
